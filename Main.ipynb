# Spectral Signature Continuum Removal - Object-Oriented Design

This notebook uses the `ContinuumRemoval` class from `continuum_removal.py` for continuum removal of spectral signatures using convex hull computation.

## ContinuumRemoval Class

- **Location**: `continuum_removal.py` (external module)
- **Purpose**: Remove continuum from spectral signatures using the convex hull method
- **Key Methods**:
  - `compute_convex_hull()`: Computes the convex hull continuum for the spectral signature
  - `compute_continuum_removed()`: Computes the continuum-removed spectrum
  - `plot_analysis(title)`: Plots the original spectrum, convex hull, and continuum-removed results

## Usage

```python
from continuum_removal import ContinuumRemoval

continuum_removal = ContinuumRemoval(wavelength, spectral_signature)
continuum_removal.plot_analysis(title="Your Title")
```


````python
from scipy.io import loadmat
import matplotlib.pyplot as plt
import numpy as np

# Import the ContinuumRemoval class from the external module
from continuum_removal import ContinuumRemoval
````

````python
# Load your data
GT = loadmat("groundTruth_Cuprite_end12/groundTruth_Cuprite_nEnd12.mat")
wavelength = np.squeeze(GT["waveLength"])
M = GT["M"]
````

````python
print(GT.keys())
````

````python
# Process each spectral sample using OOP approach
for i in range(M.shape[1]):
    spectral_signature = M[:, i]
    
    # Create ContinuumRemoval instance for this spectrum
    continuum_removal = ContinuumRemoval(wavelength, spectral_signature)
    
    # Get coordinate name for the title
    title = f"Spectrum of {GT['cood'][i][0][0]}"
    
    # Compute and plot continuum removal results
    continuum_removal.plot_analysis(title=title)
````
